```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
dataDC = pd.DataFrame(
    {
        "Образцовый вольтметр DC": 
        np.array([1.0003, 2.0004, 2.9999, 3.9991, 4.9994, 5.9992, 6.9999, 7.9996, 8.9984, 9.9988]),
        "Проверяемый вольтметр DC":
        np.arange(1, 11, 1)
    }
)
```

```python

dataAC = pd.DataFrame(
    {
        "Образцовый вольтметр AC": 
        np.array([0.9973, 1.9957, 2.9975, 3.9993, 5.0011, 6.0008, 6.9994]),
        "Проверяемый вольтметр AC":
        np.arange(1, 8, 1)
    }
)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(20, 20))

ax[0, 0].scatter(dataDC["Образцовый вольтметр DC"], dataDC["Проверяемый вольтметр DC"], c="red", label="Показания DC")
ax[0, 0].plot(np.arange(1, 11, 1), np.arange(1, 11, 1), lw=0.5, label="Образцовый вольтметр")
ax[0, 0].grid()
ax[0, 0].legend()
ax[0, 0].set_title("DC значения")



ax[0, 1].plot(np.arange(1, 11, 1), (-dataDC["Образцовый вольтметр DC"] + dataDC["Проверяемый вольтметр DC"]))
ax[0, 1].scatter(np.arange(1, 11, 1), (-dataDC["Образцовый вольтметр DC"] + dataDC["Проверяемый вольтметр DC"]), label="Абсолютная погрешность")
ax[0, 1].set_title("$\Delta U_{DC} = U_x - U$")
ax[0, 1].grid()
ax[0, 1].legend()


ax[1, 0].scatter(dataAC["Образцовый вольтметр AC"], dataAC["Проверяемый вольтметр AC"], c="red", alpha=0.9, label="Показания AC")
# ax[1, 0].scatter(dataAC["Образцовый вольтметр AC"], dataAC["Образцовый вольтметр AC"], c="blue", alpha=0.9)
ax[1, 0].plot(np.arange(1, 8, 1), np.arange(1, 8, 1), lw=0.5, label="Образцовый вольтметр")
ax[1, 0].set_title("AC значения")
ax[1, 0].grid()
ax[1, 0].legend()




ax[1, 1].plot(np.arange(1, 8, 1), (-dataAC["Образцовый вольтметр AC"] + dataAC["Проверяемый вольтметр AC"]))
ax[1, 1].scatter(np.arange(1, 8, 1), (-dataAC["Образцовый вольтметр AC"] + dataAC["Проверяемый вольтметр AC"]), label="Абсолютная погрешность")
ax[1, 1].set_title("$\Delta U_{AC} = U_x - U$")
ax[1, 1].grid()
ax[1, 1].legend()



dataDC["Разница между проверяемыми и образцовыми показаниями DC"] = dataDC.iloc[:, 1] - dataDC.iloc[:, 0]
dataAC["Разница между проверяемыми и образцовыми показаниями AC"] = dataAC.iloc[:, 1] - dataAC.iloc[:, 0]

plt.savefig("lab2_graphics.png")
```

```python
df = pd.concat([dataDC, dataAC], axis=1)
```
